Indicar titulos de categoria 'Men' para los cuales el importe de las ventas de los productos supera el promedio mensual de la misma categoria

```python
sales = [
    [1, 1, 2019, 1, 20],
    [1, 1, 2019, 2, 12],
    [2, 1, 2019, 1, 34],
    [2, 1, 2019, 3, 5],
    [3, 1, 2019, 1, 20],
    [3, 2, 2019, 2, 13],
    [4, 2, 2019, 4, 8],
    [5, 1, 2019, 3, 25],
    [1, 2, 2019, 2, 19],
    [1, 2, 2019, 1, 21],
    [2, 1, 2019, 3, 5],
    [2, 1, 2019, 5, 30],
    [2, 1, 2019, 5, 31],
    [2, 1, 2019, 6, 23],
    [2, 1, 2019, 6, 25],
    [2, 1, 2019, 8, 17],
    [2, 1, 2019, 7, 50],
]

df = pd.DataFrame(sales, columns = ['dia', 'mes', 'anio', 'product_id', 'precio'])
```

```python
descripcion = [
    [1, 'titulo1', 'men', 's/d'],
    [2, 'titulo2', 'women', 's/d'],
    [3, 'titulo3', 'kids', 's/d'],
    [4, 'titulo4', 'men', 's/d'],
    [5, 'titulo5', 'men', 's/d'],
    [6, 'titulo6', 'men', 's/d'],
    [7, 'titulo7', 'woman', 's/d'],
    [8, 'titulo8', 'kids', 's/d'],
]

ds = pd.DataFrame(descripcion, columns = ['product_id', 'titulo', 'category', 'descripcion'])
```

```python
#precio_prom = df.groupby(['mes', 'product_id']).agg({'precio': 'mean'}).reset_index()
#precio_prom.columns = ['mes', 'product_id', 'prom_mes_product']
```

```python
join = df.merge(ds.loc[:, ['product_id', 'titulo', 'category']], on='product_id', how='left')
join
```

```python
mens = join.loc[join['category'] == 'men']\

mens['promedio_mes'] = mens.groupby('mes')['precio'].transform('mean')

mens
```

```python
precio_mayor = mens.loc[mens['promedio_mes'] < mens['precio']]
precio_mayor.drop_duplicates()
```
