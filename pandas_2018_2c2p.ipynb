```python
mens = ds.loc[ds['category'] == 'men'] 

join = df[['mes', 'product_id', 'precio']].merge(mens.loc[:, ['product_id', 'titulo', 'category']], on='product_id', how='inner')
join
```

```python
# mens = join.loc[join['category'] == 'men']\

join['promedio_mes'] = join.groupby('mes')['precio'].transform('mean')

join
```

```python
precio_mayor = join.loc[join['promedio_mes'] < join['precio']]
precio_mayor.drop_duplicates()
```

```python
df['fecha'] = (df['anio'] + df['mes'] + df['dia']).astype(int)
df.drop(columns=['anio', 'mes', 'dia'], inplace=True)
df.head()
```

```python
def to_listo(x):
    fechas_list = sorted(set(x))
    fechas_list.append(20190219)
    consecutivos = []
    consec=0
    j=1
    for i in range(0, len(fechas_list)-1):
        if fechas_list[j] == fechas_list[j-1]+1:
            consecutivos.append(fechas_list[consec])
            consec+=1
        j+=1
    return consecutivos
```

```python
list_fechas = pd.DataFrame(df.groupby('product_id')['fecha'].apply(to_listo).reset_index())
list_fechas
```
